**Calibration score**

```python
def S_vector(realization, Expert):
    '''
    Calculate the vector s of an expert's answer
    
    Inputs : 
    realization - vector with answers to the questions
    Expert - matrix with experts quantiles 5, 50, 95
    
    Outputs : 
    s_solution - vector s of an expert
    '''
    s_empty = np.zeros(4)
    
    s_empty[0] = np.sum(np.less_equal(realization, Expert[:,0])) # count where realization lies
    s_empty[1] = np.sum(np.less(Expert[:,0], realization) & np.less_equal(realization, Expert[:,1]))
    s_empty[2] = np.sum(np.less(Expert[:,1], realization) & np.less_equal(realization, Expert[:,2]))
    s_empty[3] = np.sum(np.less_equal(Expert[:,2], realization))
    
    s_solution = s_empty / len(realization) # normalization
    
    return s_solution
```

```python
def KLD(s_vector):
    '''
    Calculate the KL Divergence used for the calculation of the calibration score
    
    Inputs : 
    s_vector - vector s of the expert
    
    Outputs :s
    I - Index used inside Chi squared for the calibration score
    '''
    p_vector = [0.05, 0.45, 0.45, 0.05]
    
    find_0 = np.where(s_vector == 0) # convention : if s = 0 , delete it
    s_vector = np.delete(s_vector, find_0)
    p_vector = np.delete(p_vector, find_0)
    
    I = np.sum(s_vector * np.log(s_vector / p_vector))

    return I

def Calibration_coef(m, s_vector):
    '''
    Calculate the calibration score of an expert
    
    Inputs : 
    s_vector - vector s of the expert
    m - number of questions
    
    Outputs : 
    Cal - Calibration score
    '''
    
    Cal = 1 - chi2.cdf(2 * m * KLD(s_vector), 3, loc=0, scale=1)
    
    return Cal
```

**Information score**

```python
def U_L (realization, k, *args):
    '''
    Calculate the vectors U* and L* of a group of experts. 
    
    Inputs : 
    number of questions
    realization - vector with answers to the questions or -9.99E99 for questions of interest
    *args - experts involves in the case (an arbitrary number of experts can be introduced)
    
    Outputs : 
    L_vector - vector of bottom limit values for each question
    U_vector - vector of top limit values for each question
    '''
    minimum = []
    maximum = []
    U = np.zeros(len(realization))
    L = np.zeros(len(realization))
    
    for j in range(len(realization)):
        if realization[j] == -9.99E99:
            maximum = maximum
            minimum = minimum
        
        else: 
            maximum.append(realization[j])
            minimum.append(realization[j])
        
        for i in range(len(args)):
            minimum.append(args[i][j, 0]) # take from each expert each question quantile 5 & 95
            maximum.append(args[i][j, 2])

        U[j] = np.max(maximum)
        L[j] = np.min(minimum)
        
        minimum = []
        maximum = []
        
    L_vector = L - k * (U-L) 
    U_vector = U + k * (U-L)
        
    return L_vector, U_vector
```

```python
def Information_coef(Expert, U_vector, L_vector):
    '''
    Calculate information vector with coefficients for each question for an expert.
    In case a realization is not given, 
    the experts values can be introduced as the realization as it woulnd't modify the vector at all.
    
    Inputs : 
    realization - vector with answers to the questions
    Expert - matrix with experts quantiles 5, 50, 95
    L_vector - vector of bottom limit values for each question
    U_vector - vector of top limit values for each question
    
    Outputs : 
    Information_Expert - vector with information scores for each question
    '''     
    Information_Expert = (0.05 * np.log(0.05 / (Expert[:, 0]-L_vector)) + 
                          0.45 * np.log(0.45 / (Expert[:, 1]-Expert[:, 0])) + 
                          0.45 * np.log(0.45 / (Expert[:, 2]-Expert[:, 1])) + 
                          0.05 * np.log(0.05 / (U_vector-Expert[:, 2])) + 
                          np.log(U_vector - L_vector)) 
    
    return Information_Expert
```
